```python
import random
board=["__"]*9
player="X"
computer="0"
def display_board(board):
    print(board[0]+"|"+board[1]+"|"+board[2])
    print("-"*9)
    print(board[3]+"|"+board[4]+"|"+board[5])
    print("-"*9)
    print(board[6]+"|"+board[7]+"|"+board[8])
    print('________________')
def check_win(board):
    if board[0]==board[1]==board[2] and board[0]!="__":
        return True
    elif board[3]==board[4]==board[5] and board[5]!="__":
        return True
    elif board[6]==board[7]==board[8] and board[6]!="__":
        return True
    elif board[0]==board[3]==board[6] and board[0]!="__":
        return True
    elif board[1]==board[4]==board[7] and board[1]!="__":
        return True
    elif board[2]==board[5]==board[8] and board[2]!="__":
        return True
    elif board[0]==board[4]==board[8] and board[0]!="__":
        return True
    elif board[2]==board[4]==board[6] and board[2]!="__":
        return True
    else:
        return False
def check_tie(board):
        return not check_win(board) and "__" not in board
def is_available(position):
    return board[position]=="__"
      
def insert(symbol,position):
    if is_available(position):
        board[position]=symbol
        display_board(board)
        if check_win(board):
            if symbol=="0":
                print ("computer wins")
            else:
                print("Player wins")
            exit()
        if check_tie(board):
            print("it's a Tie")
            exit()
    else:
        if symbol==player:
            position=int(input("Not free ! please re enter a position:"))
        else:
            position=random.randint(0,8)
        insert(symbol,position)
def player_move(symbol):
    position=int(input("enter the position to insert :"))
    insert(symbol,position)
def computer_move(symbol):
    position=random.randint(0,8)
    insert(symbol,position)
while True:
    display_board(board)
    player_move(player)
    if check_win(board) or check_tie(board):
        break
    computer_move(computer)
    if check_win(board) or check_tie(board):
        break
   
    
print("end game")
    
```
