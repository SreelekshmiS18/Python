```python
import random

class HandCricket:#define class
    def __init__(self):
        self.player_runs = 0
        self.ai_runs = 0
        self.player_name = ""
        self.game_over = False

    def play_again(self):
        again = input("Would you like to play again (yes/no): ")
        if again.upper() == "YES":
            self.reset_game()
            self.main()
        elif again.upper() == "NO":
            print('Thank you for playing!')
            exit()
        else:
            print("Invalid input. Exiting.")
            exit()

    def reset_game(self):
        self.player_runs = 0
        self.ai_runs = 0
        self.game_over = False

    def first_batting(self):
        print("You are going to bat first.")
        while not self.game_over:
            me = int(input("Bat: "))
            AI = random.randint(1, 10)
            print("AI ball: ", AI)
            self.player_runs += me
            print("Your current score is: ", self.player_runs)
            if me == AI:
                print("AI took your wicket!")
                print("AI is going to bat now.")
                print("AI needs", (self.player_runs + 1), "runs to win the match.")
                self.ai_batting()

    def ai_batting(self):
        while not self.game_over:
            me = int(input("Ball: "))
            AI = random.randint(1, 10)
            print("AI bat: ", AI)
            self.ai_runs += AI
            print("The current score of AI is: ", self.ai_runs)
            if self.ai_runs > self.player_runs:
                print("Good try buddy, but AI won the match. Better luck next time!")
                self.game_over = True
                self.play_again()

    def first_bowling(self):
        print("You are going to bowl first.")
        while not self.game_over:
            me = int(input("Ball: "))
            AI = random.randint(1, 10)
            print("AI bat: ", AI)
            self.ai_runs += AI
            print("Current score of AI is: ", self.ai_runs)
            if me == AI:
                print("You took the wicket of AI!")
                print("AI is going to bowl now.")
                print("You need", (self.ai_runs + 1), "runs to win the match.")
                self.player_batting()

    def player_batting(self):
        while not self.game_over:
            me = int(input("Bat: "))
            AI = random.randint(1, 10)
            print("AI bat: ", AI)
            self.player_runs += me
            print("Your current score is: ", self.player_runs)
            if self.player_runs > self.ai_runs:
                print("You won the match!")
                self.game_over = True
                self.play_again()

    def main(self):
        print("Welcome to the game of hand cricket in Python!")
        self.player_name = input("Please enter your name: ")
        toss = input("Enter ODD or EVEN: ").upper()
        me = int(input("Enter any number between 1 to 10: "))

        
        if me < 1 or me > 10:
            print("Invalid input. Please enter a number between 1 and 10.")
            return

        AI = random.randint(1, 10)
        print("Input of AI: ", AI)

        if toss == "EVEN":
            if (me + AI) % 2 == 0:
                print("You won the toss. Choose to bat or bowl.")
                toss_input = input("BAT or BALL: ").upper()
                if toss_input == "BAT":
                    self.first_batting()
                elif toss_input == "BALL":
                    self.first_bowling()
                else:
                    print("Invalid input.")
                    return
            else:
                print("AI wins the toss.")
                AItoss = random.choice(["BAT", "BALL"])
                if AItoss == "BAT":
                    print("AI is going to bat first.")
                    self.first_bowling()
                else:
                    print("AI is going to bowl first.")
                    self.first_batting()

        elif toss == "ODD":
            if (me + AI) % 2 != 0:
                print("You won the toss. Choose to bat or bowl.")
                toss_input = input("BAT or BALL: ").upper()
                if toss_input == "BAT":
                    self.first_batting()
                elif toss_input == "BALL":
                    self.first_bowling()
                else:
                    print("Invalid input.")
                    return
            else:
                print("AI wins the toss.")
                AItoss = random.choice(["BAT", "BALL"])
               
                if AItoss == "BAT":
                    print("AI is going to bat first.")
                    self.first_bowling()
                else:
                    print("AI is going to bowl first.")
                    self.first_batting()

if __name__ == "__main__":
    game = HandCricket()
    game.main()
```
